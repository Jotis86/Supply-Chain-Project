```python
# Import librearies for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load the data
data = pd.read_csv('cleaned_data.csv')
```

```python
# Display the first 5 rows of the data
data.head()
```

```python
# Convert date columns to datetime
data['fecha_odc'] = pd.to_datetime(data['fecha_odc'])
data['fecha_entrega'] = pd.to_datetime(data['fecha_entrega'])
data['fecha_recibido'] = pd.to_datetime(data['fecha_recibido'])


# Display the information of the data
data.info()
```

```python
# 1. Histogram of ordered product quantity
plt.figure(figsize=(12, 8))
sns.set_style(style="whitegrid")

# Create the histogram
sns.histplot(data['cant_prod_odc'], bins=30, kde=True, color='skyblue')

# Add titles and labels
plt.title('Histogram of Ordered Product Quantity', fontsize=16)
plt.xlabel('Ordered Product Quantity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

```python
# 2. Histogram of unit price
plt.figure(figsize=(12, 8))

# Create the histogram
sns.histplot(data['prec_unt'], bins=30, kde=True, color='salmon', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Unit Price', fontsize=16, fontweight='bold')
plt.xlabel('Unit Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# 3. Histogram of order amount
plt.figure(figsize=(12, 8))

# Create the histogram
sns.histplot(data['monto_odc'], bins=30, kde=True, color='lightgreen', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Order Amount', fontsize=16, fontweight='bold')
plt.xlabel('Order Amount', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# 4. Histogram of received quantity
plt.figure(figsize=(12, 8))

# Create the histogram
sns.histplot(data['cant_recibida'], bins=30, kde=True, color='lightblue', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Received Quantity', fontsize=16, fontweight='bold')
plt.xlabel('Received Quantity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# 5. Histogram of received amount
plt.figure(figsize=(12, 8))

# Create the histogram
sns.histplot(data['monto_recibido'], bins=30, kde=True, color='purple', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Received Amount', fontsize=16, fontweight='bold')
plt.xlabel('Received Amount', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# 6. Bar plot of order status
plt.figure(figsize=(12, 8))


# Create the count plot
sns.countplot(data=data, x='estado_odc', palette='viridis', edgecolor='black')

# Add titles and labels
plt.title('Order Status Distribution', fontsize=18, fontweight='bold')
plt.xlabel('Order Status', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# 7. Line plot of order amount over time
plt.figure(figsize=(12, 8))

# Create the line plot
sns.lineplot(data=data, x='fecha_odc', y='monto_odc', marker='o', color='b', linewidth=2.5)

# Add titles and labels
plt.title('Order Amount Over Time', fontsize=18, fontweight='bold')
plt.xlabel('Order Date', fontsize=14)
plt.ylabel('Order Amount', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(4, 2, figsize=(20, 20))
fig.suptitle('Dashboard Summary', fontsize=20)

# 1. Histogram of ordered product quantity
sns.histplot(data['cant_prod_odc'], bins=30, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Histogram of Ordered Product Quantity', fontsize=15)
axes[0, 0].set_xlabel('Ordered Product Quantity', fontsize=12)
axes[0, 0].set_ylabel('Frequency', fontsize=12)

# 2. Histogram of unit price
sns.histplot(data['prec_unt'], bins=30, kde=True, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Histogram of Unit Price', fontsize=15)
axes[0, 1].set_xlabel('Unit Price', fontsize=12)
axes[0, 1].set_ylabel('Frequency', fontsize=12)

# 3. Line plot of order amount over time
sns.lineplot(data=data, x='fecha_odc', y='monto_odc', ax=axes[1, 0], color='coral')
axes[1, 0].set_title('Order Amount Over Time', fontsize=15)
axes[1, 0].set_xlabel('Order Date', fontsize=12)
axes[1, 0].set_ylabel('Order Amount', fontsize=12)

# 4. Box plot of unit price by order status
sns.boxplot(data=data, x='estado_odc', y='prec_unt', ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title('Unit Price by Order Status', fontsize=15)
axes[1, 1].set_xlabel('Order Status', fontsize=12)
axes[1, 1].set_ylabel('Unit Price', fontsize=12)

# 5. Scatter plot of ordered quantity vs received quantity
sns.scatterplot(data=data, x='cant_prod_odc', y='cant_recibida', ax=axes[2, 0], color='purple')
axes[2, 0].set_title('Ordered Quantity vs Received Quantity', fontsize=15)
axes[2, 0].set_xlabel('Ordered Quantity', fontsize=12)
axes[2, 0].set_ylabel('Received Quantity', fontsize=12)

# 6. Bar plot of delivery days distribution (limited to 30 days)
sns.histplot(data[data['delivery_days'] <= 30]['delivery_days'], bins=30, kde=True, ax=axes[2, 1], color='gold')
axes[2, 1].set_title('Distribution of Delivery Days (<= 30)', fontsize=15)
axes[2, 1].set_xlabel('Delivery Days', fontsize=12)
axes[2, 1].set_ylabel('Frequency', fontsize=12)

# 7. Bar plot of on-time deliveries
sns.countplot(data=data, x='ontime', ax=axes[3, 0], palette='Set2')
axes[3, 0].set_title('On-time Deliveries', fontsize=15)
axes[3, 0].set_xlabel('On-time', fontsize=12)
axes[3, 0].set_ylabel('Count', fontsize=12)

# 8. Bar plot of OTIF deliveries
sns.countplot(data=data, x='OTIF', ax=axes[3, 1], palette='Set3')
axes[3, 1].set_title('OTIF Deliveries', fontsize=15)
axes[3, 1].set_xlabel('OTIF', fontsize=12)
axes[3, 1].set_ylabel('Count', fontsize=12)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
