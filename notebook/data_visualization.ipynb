```python
# Import librearies for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load the data
data = pd.read_csv('cleaned_data.csv')
```

```python
# Display the first 5 rows of the data
data.head()
```

```python
# Convert date columns to datetime
data['fecha_odc'] = pd.to_datetime(data['fecha_odc'])
data['fecha_entrega'] = pd.to_datetime(data['fecha_entrega'])
data['fecha_recibido'] = pd.to_datetime(data['fecha_recibido'])


# Display the information of the data
data.info()
```

```python
# 1. Histogram of ordered product quantity
plt.figure(figsize=(12, 8))
sns.set_style(style="whitegrid")

# Create the histogram
sns.histplot(data['cant_prod_odc'], bins=30, kde=True, color='skyblue')

# Add titles and labels
plt.title('Histogram of Ordered Product Quantity', fontsize=16)
plt.xlabel('Ordered Product Quantity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

```python
# 2. Histogram of unit price
plt.figure(figsize=(12, 8))

# Create the histogram
sns.histplot(data['prec_unt'], bins=30, kde=True, color='salmon', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Unit Price', fontsize=16, fontweight='bold')
plt.xlabel('Unit Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(5, 2, figsize=(20, 25))
fig.suptitle('Dashboard Summary', fontsize=20)

# 1. Bar plot of order status
sns.countplot(data=data, x='estado_odc', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Order Status Distribution', fontsize=15)
axes[0, 0].set_xlabel('Order Status', fontsize=12)
axes[0, 0].set_ylabel('Count', fontsize=12)

# 2. Bar plot of origin country
sns.countplot(data=data, x='org_pais', ax=axes[0, 1], palette='magma')
axes[0, 1].set_title('Origin Country Distribution', fontsize=15)
axes[0, 1].set_xlabel('Origin Country', fontsize=12)
axes[0, 1].set_ylabel('Count', fontsize=12)

# 3. Box plot of unit price by category
sns.boxplot(data=data, x='Categoria', y='prec_unt', ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title('Unit Price by Category', fontsize=15)
axes[1, 0].set_xlabel('Category', fontsize=12)
axes[1, 0].set_ylabel('Unit Price', fontsize=12)
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Box plot of unit price by subcategory
sns.boxplot(data=data, x='Subcategoria', y='prec_unt', ax=axes[1, 1], palette='coolwarm')
axes[1, 1].set_title('Unit Price by Subcategory', fontsize=15)
axes[1, 1].set_xlabel('Subcategory', fontsize=12)
axes[1, 1].set_ylabel('Unit Price', fontsize=12)
axes[1, 1].tick_params(axis='x', rotation=45)

# 5. Scatter plot of cost vs unit price
sns.scatterplot(data=data, x='costo_prod', y='prec_unt', ax=axes[2, 0], color='darkblue', s=100, alpha=0.6)
axes[2, 0].set_title('Cost vs Unit Price', fontsize=15)
axes[2, 0].set_xlabel('Cost', fontsize=12)
axes[2, 0].set_ylabel('Unit Price', fontsize=12)

# 6. Scatter plot of delivery days vs reception days
sns.scatterplot(data=data, x='delivery_days', y='reception_days', ax=axes[2, 1], color='darkgreen', s=100, alpha=0.6)
axes[2, 1].set_title('Delivery Days vs Reception Days', fontsize=15)
axes[2, 1].set_xlabel('Delivery Days', fontsize=12)
axes[2, 1].set_ylabel('Reception Days', fontsize=12)

# 7. Line plot of total amount over time
sns.lineplot(data=data, x='fecha_odc', y='total_amount', ax=axes[3, 0], color='coral', linewidth=2.5)
axes[3, 0].set_title('Total Amount Over Time', fontsize=15)
axes[3, 0].set_xlabel('Order Date', fontsize=12)
axes[3, 0].set_ylabel('Total Amount', fontsize=12)

# 8. Bar plot of total amount by category
category_sales = data.groupby('Categoria')['total_amount'].sum().reset_index()
sns.barplot(data=category_sales, x='Categoria', y='total_amount', ax=axes[3, 1], palette='Blues_d')
axes[3, 1].set_title('Total Amount by Category', fontsize=15)
axes[3, 1].set_xlabel('Category', fontsize=12)
axes[3, 1].set_ylabel('Total Amount', fontsize=12)
axes[3, 1].tick_params(axis='x', rotation=45)

# 9. Bar plot of on-time deliveries by provider
sns.countplot(data=data, x='nom_prov', hue='ontime', ax=axes[4, 0], palette='Set2')
axes[4, 0].set_title('On-time Deliveries by Provider', fontsize=15)
axes[4, 0].set_xlabel('Provider', fontsize=12)
axes[4, 0].set_ylabel('Count', fontsize=12)
axes[4, 0].tick_params(axis='x', rotation=90)
axes[4, 0].legend(title='On-time', fontsize=12)

# 10. Bar plot of OTIF deliveries by provider
sns.countplot(data=data, x='nom_prov', hue='OTIF', ax=axes[4, 1], palette='Set3')
axes[4, 1].set_title('OTIF Deliveries by Provider', fontsize=15)
axes[4, 1].set_xlabel('Provider', fontsize=12)
axes[4, 1].set_ylabel('Count', fontsize=12)
axes[4, 1].tick_params(axis='x', rotation=90)
axes[4, 1].legend(title='OTIF', fontsize=12)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
